### Задача 1. Визуализация средних, заливка области между графиками

Отобразите на одном графике:
* область между `Low` и `High`;
* плавающее среднее для `(Low + High) / 2` с 2-недельным окном;
* среднее в экспоненциально взвешенном плавающем окне для `(Low + High) / 2` с 3-дневным периодом полураспада.

Для повышения читаемости выведите легенду и сетку.

#### Подсказки
* Легенда выводится с помощью функции `plt.legend`, сетка — с помощью `plt.grid`.
* Для отображения области между графиками используйте [функцию `plt.fill_between`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.fill_between).
* Вычисление необходимых средних осуществляется с помощью [методов `rolling` и `ewm`](http://pandas.pydata.org/pandas-docs/stable/computation.html?highlight=rolling#window-functions). Не забудьте применить агрегатор (в данном случае `mean`) к результату соответствующей функции.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start_date = '20160601'
end_date = '20160830'
plt.figure(figsize=(15, 5))
plt.grid()

fbl = fb.loc[start_date:end_date]
plt.fill_between(fbl.index, fbl.Low, fbl.High, lw=0, color='b', alpha=0.3, label='[Low, High]')

fb_roll = fb.rolling(window=10).mean()
fb_roll = fb_roll[fb_roll.Open.notnull()].loc[start_date:end_date]
plt.plot(fb_roll.index, (fb_roll.Low + fb_roll.High) / 2, color='r', lw=2, label='Rolling mean')

fb_ewm = fb.ewm(halflife=3).mean()[start_date:end_date]
plt.plot(fb_ewm.index, (fb_ewm.Low + fb_ewm.High) / 2, color='g', lw=2, label='EW mean')

plt.legend(loc='upper left')
```

### Задача 2. Вывод дополнительной информации с помощью error bars

С помощью функции `plt.errorbar` выведите график цены (например, `(Low + High) / 2`), снабдив его дополнительной информацией об объёмах сделок.

### Задача 3.1. Построение свечного графика

Постройте свечной график для указанных выше данных. Окрасьте свечи в красный (`'red'`) и ярко-зелёный (`'lime'`) цвета.

#### Подсказки

* Используйте [функцию `candlestick2_ohlc` из модуля `matplotlib.finance`](http://matplotlib.org/api/finance_api.html#matplotlib.finance.candlestick2_ohlc).

### Задача 3.2. Совмещение свечного графика с информацией об объёмах сделок

С помощью функции `plt.subplots` создайте два графика, имеющих общую ось _x_. На верхнем выведите свечной график, на втором — объёмы сделок с помощью метода `bar`. Верхний график должен значительно превышать по размеру второй.
Добавьте подписи _y_-осям графиков.

#### Подсказки

* Использование функции `plt.subplots` для создания подграфиков: [Easily creating subplots](http://matplotlib.org/users/recipes.html?highlight=subplots#easily-creating-subplots). [Примеры](http://matplotlib.org/examples/pylab_examples/subplots_demo.html).
* Для получения подграфиков разных размеров используйте параметр `gridspec_kw` функции `plt.subplots`. [Пример](http://stackoverflow.com/a/35881382).
* Добавить подписи осям можно с помощью метода `set_ylabel` объектов `ax`.

```python
fig, (ax1, ax2) = plt.subplots(
    2, 1,
    sharex=True, sharey=False,
    figsize=(15, 7), gridspec_kw={'height_ratios': [5, 1]}
)

ax1.grid()
ax1.set_xlim((-1, len(fbl)))
ax1.set_ylabel('FB price')
candlestick2_ohlc(ax1, fbl.Open, fbl.High, fbl.Low, fbl.Close, colorup='lime', colordown='r', width=.8)

ax2.set_ylabel('FB volume')
ax2.bar(range(len(fbl)), np.array(fbl.Volume), width=1, align='center', lw=1, color='g', alpha=0.4)
```

### Задача 3.3 (на любителя). Кастомизация значений на осях

### Задача 4. Класс `pandas.HDFStore` для работы с таблицами на диске

#### Подготовка данных: быстрая генерация небольшого файла со случайными данными

```python
def build_file_row(timestamp, price, volume):
    return '{timestamp},{price:.2f},{volume}\n'.format(timestamp=timestamp, price=price, volume=volume)

def create_small_file(filename, rows=10**3):
    file = open(filename, 'w', encoding='cp1251')
    file.write('Timestamp,Price,Volume\n')
    
    timestamp = dt.datetime(2010, 1, 1)
    price = 200
    
    for i in range(rows):
        volume = int(np.random.exponential(scale=10) + 1)
        
        file.write(build_file_row(timestamp, price, volume))
        
        timestamp += dt.timedelta(seconds=1)
        price += np.random.normal() / 10
        price = max(price, 10)
    file.close()
    
create_small_file('finam-6-smallfile.csv', rows=10**4)
```

#### Создание файла `.h5`

Считав данные из `finam-6-smallfile.csv` кусками по 1000 строк, создайте файл `finam-6-smallfile.h5`, содержащий всю таблицу.

```python
import pandas as pd
store = pd.HDFStore('finam-6-smallfile.h5', mode='w')
for chunk in pd.read_csv('finam-6-smallfile.csv', parse_dates=['Timestamp'], index_col='Timestamp', chunksize=1000):
    store.append('df', chunk, data_columns=True)
store.close()
```

#### Загрузка данных из таблицы на диске по условию

Загрузите из полученного файла все данные до 01.01.2010 02:00.
